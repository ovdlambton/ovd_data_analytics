```python
import time
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# User enters the number of terms to generate in the Fibonacci series
#n = int(input("Enter the number of terms: "))
n = 35 # changed to fixed value to ease checking of the program, to check other terms use the command above instead
```

```python
# Recursive method to generate nth Fibonacci number based on the terms entered by the user
def fibonacci_r(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_r(n-1) + fibonacci_r(n-2)

print("Recursive Fibonacci Series Result Below:")
for i in range(n+1):
    print(fibonacci_r(i), end=", ")
```

```python
# Non-recursive method to generate nth Fibonacci number based on the terms entered by the user
print("Non Recursive Fibonacci Series Result Below:")

def fibonacci_nr(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for i in range(2,n+1):
            c = a + b
            a = b
            b = c
        return b

for i in range(n+1):
    print(fibonacci_nr(i), end=", ")
```

```python
# Create lists to store the execution times and Fibonacci numbers
recursive_times = []
recursive_numbers = []
non_recursive_times = []
non_recursive_numbers = []
fib_series_numbers = pd.Series(np.arange(n+1))

#print(fib_series_numbers)
```

```python
# Generate the Fibonacci series using the recursive method and store execution time
for i in range(n+1):
    start_time = time.time_ns() # Store start time of each iteration in nanoseconds
    number = fibonacci_r(i)
    end_time = time.time_ns() # Store end time of each iteration in nanoseconds
    elapsed_time = end_time - start_time # Calculates the elapsed time for each iteration
    recursive_times.append(elapsed_time)
    recursive_numbers.append(number)
```

```python
# Generate the Fibonacci series using the non-recursive method and store execution time
for i in range(n+1):
    start_time = time.time_ns() # Store start time of each iteration in nanoseconds
    number = fibonacci_nr(i)
    end_time = time.time_ns() # Store end time of each iteration in nanoseconds
    elapsed_time = end_time - start_time # Calculate elapsed time for each iteration
    non_recursive_times.append(elapsed_time)
    non_recursive_numbers.append(number)
```

```python
# Plot the Fibonacci numbers against the execution times for the recursive method
plt.plot(fib_series_numbers, recursive_times, label='Recursive')
# Plot the Fibonacci numbers against the execution times for the non-recursive method
plt.plot(fib_series_numbers, non_recursive_times, label='Non-recursive')

# Add labels and legend to the plot
plt.xlabel('Fibonacci Number')
plt.ylabel('Execution Time (nanoseconds)')
plt.legend()

# Show the plot
plt.show()
```

Recursion involves creating a function that calls itself while with Non-Recursion it involves iterative loop structure to perform the calculation. The Recursion approach allows for a simpler algorithm, but it has a certain downside as it takes longer to run since each number in the sequence is computed twice by calling itself (the individual numbers). While the Non-Recursion approach requires more complex code but can be computationally quicker since only each Fibonacci number needs to be calculated once.
